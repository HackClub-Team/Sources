**If Elif Else**
* **If**
```
if some_condition:
     # execute some code
```

* **If Else**
```
if some_condition:
     # execute some code
else:
     # do something else
```

* **Elif**
```
if some_condition:
     # execute some code
elif some_other_condition:
     # do something different
else:
     # do something else
```

````python
if 3 > 2:
  print("3 is greater than 2")
````

````python
hungry = False

if hungry:
  print("I am hungry")
else:
  print("I am not hungry")
````

````python
something = "Bank"

if something == "Auto Shop":
  print("Cars are cool")
elif something == "Bank":
  print("Money is cool")
elif something == "Store":
  print("Welcome to the store")
else:
  print("Idk")
````
