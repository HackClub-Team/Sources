**Useful Operators**

```python
for num in range(10):
  print(num)
```

```python
for num in range(3, 10, 2):
  print(num)
```

```python
list(range(0, 11, 2))
```

```python
index_count = 0

for letter in "abcde":
  print("At index {} the letter is {}".format(index_count, letter))
  index_count += 1
```

```python
for item in enumerate("abcde"):
  print(item)
```

```python
my_list1 = [1, 2, 3]
my_list2 = ["a", "b", "c"]
my_list3 = [100, 200, 300]

for item in zip(my_list1, my_list2, my_list3):
  print(item)
```
