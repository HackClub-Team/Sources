**Variables**
* Can't use any of these symbols
> :"',<>/?|\()!@#$%^&*~-+
* Good practice to keep variable names in lowercase, preferably snake_case
> variable_name, character_speed
* Python uses dynamic typing, so variable types can be changed anytime
> * my_dogs = 2
> * my_dogs = ["Sammy", "Frankie"]
