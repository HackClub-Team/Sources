**Strings**
* Can use single quotes or double quotes. Most programmers use double quotes because of apostrophes
> * 'Hello'
> * "Hello"
* Index starts from 0 instead of 1. Reverse index starts at 0 and goes from the final index to the first.
> * h e l l  o
> * 0 1 2 3 4
> * 0 -4 -3 -2 -1
* Slicing allows you grab a subsection of multiple characters, a "slice" of the string: **[start:stop:step]**
> * start is a numerical index for the slice start
> * stop is the index you will go up to (but not include)
> * step is the size of the "jump" you take
* Escape sequences
> * **\n** is used for a new line
> * **\t** is used for a tab
> * **\\\** is used for a backslash
* **len()** counts the number of characters in a string
* A good way to format objects into your strings for print statements is with the string **.format()** method
> * "String here { } then also { }".format("something1, "something2")
