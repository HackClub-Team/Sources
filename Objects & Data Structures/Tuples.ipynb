**Tuples**
* Tuples are very similar to lists. However, they have one key difference - **immutability**
* Once an element is inside a tuple, it can not be reassigned
* Tuples use parenthesis
> * (1, 2, 3)

```python
t = (1, 2, 3)
```

```python
my_list = [1, 2, 3]
```

```python
type(t)
```

```python
type(my_list)
```

```python
len(t)
```

```python
t = ("one", 2)
```

```python
t[0]
```

```python
t = ("a", "a", "b")
```

```python
t.count("a")
```
