**Sets**
* Sets are unordered collections of unique elements
* There can only be one representative of the same object
* Curly braces does **NOT** mean dictionary because it doesn't have any key value pairs

```python
my_set = set()
```

```python
my_set.add(1)
```

```python
my_set
```

```python
my_set.add(2)
```

```python
my_set
```

```python
my_list = [1, 1, 1, 1, 1, 1, 2, 2, 2, 3, 3, 3]
```

```python
set(my_list)
```
