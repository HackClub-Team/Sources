**Functions**
* Functions start off with a ```def``` keyword
* Generally uses snake casing
```
def name_of_function():
     # execute some code
```

* You can input arguments inside the function
```
def name_of_function(arg):
     # execute some code that uses the argument
```
* Functions can return results back
```
def name_of_function(arg):
     # execute some code
     return variable_name
```

````python
def print_name(name):
  print("Hello " + name)
````

````python
print_name("John")
````

````python
def add_numbers(num1, num2):
  return num1 + num2
````

````python
result = add_numbers(5, 6)
print(result)
````

````python
def say_hello():
  print("Hello")
````

````python
say_hello()
````

````python
def even_check(number):
  return number % 2 == 0
````

````python
even_check(20)
````

````python
even_check(21)
````

````python
def check_even_list(num_list):
  even_numbers = []

  for number in num_list:
    if number % 2 == 0:
      even_numbers.append(number)
  return even_numbers
````

````python
check_even_list([1, 3, 5])
````

````python
check_even_list([2, 4, 5])
````
